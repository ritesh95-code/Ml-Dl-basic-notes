```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['State']=le.fit_transform(df.State)
```

```python
df.head()
```

```python
X=df.iloc[:,:-1].values
```

```python
y=df.iloc[:,4:5].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
x_train.shape, x_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Creatinh ANN Seklton
regressor=Sequential()  #initialising the Model
regressor.add(Dense(4, activation='relu')) # input layer
regressor.add(Dense(8, activation='relu'))
regressor.add(Dense(16, activation='relu'))  
regressor.add(Dense(1, activation='linear')) # Output layer
```

```python
# Compiling ANN model
regressor.compile(loss='mse',optimizer='adam',metrics=['mse'])
```

```python
# Training the Model
regressor.fit(x_train,y_train, batch_size=10, epochs=300)
```

```python
# test the model
y_pred=regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
