***Question 2***

```python
# 2) Find the movie category having highest ratings in each year. The output should comprise of YearofRelease, Category, CountOfMovies having highest rating.

merged_df = new_df2.merge(new_df1, on='MovieID')
merged_df['Category'] = merged_df['Category'].apply(lambda x: ','.join(x))
# Finding highest rating categories per year
category_ratings_per_year = merged_df.groupby(['ReleaseYear', 'Category'])['Rating'].mean().reset_index()
highest_rated_category_per_year = category_ratings_per_year.loc[category_ratings_per_year.groupby('ReleaseYear')['Rating'].idxmax()]
print(highest_rated_category_per_year)
```

***Questions 3***

```python
# 3) Find and display movie category and age group wise likings. This may be based on the highest number of users from each age group rated for that movie category.

complete_df = merged_df.merge(df3, on='UserID')

# Aggregating by Category and Age Group
age_group_category_likes = complete_df.groupby(['Age', 'Category']).size().reset_index(name='Count')
most_liked_categories_per_age_group = age_group_category_likes.loc[age_group_category_likes.groupby('Age')['Count'].idxmax()]
print("\n3) Movie category and age group wise likings:")
print(most_liked_categories_per_age_group)
```

***Questions 4***

```python
# 4) Use Cluster models to segregate movie category and age group wise likings.

from sklearn.cluster import KMeans

# Preparing data for clustering
clustered_data = complete_df[['Age', 'Category']].copy()
clustered_data['Category'] = clustered_data['Category'].astype('category').cat.codes  # Encoding categories

# Running K-means clustering
kmeans = KMeans(n_clusters=2, random_state=0).fit(clustered_data)
clustered_data['Cluster'] = kmeans.labels_

print(clustered_data.groupby(['Cluster', 'Category']).size().reset_index(name='Count'))
```

***Question 7***

```python
# 7) Use Clustering methods to segregate movie category and occupation of users. Display details. With the data, train the model and predict that when an occupation is entered, the model can predict the movie likings to that category of occupation.

# Preparing data for clustering
clustered_data = complete_df[['Occupation', 'Category']].copy()
clustered_data['Category'] = clustered_data['Category'].astype('category').cat.codes  # Encoding categories

# Running K-means clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(clustered_data)
clustered_data['Cluster'] = kmeans.labels_

# Displaying clustered data
clustered_details = clustered_data.groupby(['Cluster', 'Occupation', 'Category']).size().reset_index(name='Count')
print(clustered_details)
```

```python
# Predicting Movie Likings Based on Occupation

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Preparing the dataset
x = clustered_data[['Occupation']]
y = clustered_data['Category']

# Splitting the dataset
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Training the model
model = RandomForestClassifier(random_state=42)
model.fit(xtrain, ytrain)

# Converting the 'Category' column to categorical data type
complete_df['Category'] = complete_df['Category'].astype('category')

# Predicting for a new occupation
new_occupation = [[10]]  # Example occupation
predicted_category_code = model.predict(new_occupation)
predicted_category = complete_df['Category'].cat.categories[predicted_category_code[0]]
print(f"Predicted movie category for occupation {new_occupation[0][0]}: {predicted_category}")
```
